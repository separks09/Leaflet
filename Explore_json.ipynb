```python
# Dependencies
import json
import requests 
```

```python
# Specify the URL
request_url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/2.5_week.geojson"

# Pretty Print the output of the JSON
response = requests.get(request_url)

data = response.json()

data.keys()
```

```python
quakes = []

for x in range(len(data["features"])):
    event_dict = {}
    lat = data["features"][x]["geometry"]["coordinates"][0]
    long = data["features"][x]["geometry"]["coordinates"][1]
    event_dict["location"] = [lat,long]
    event_dict["mag"] = data["features"][x]["geometry"]["coordinates"][1]
    event_dict["url"] = data["features"][x]["properties"]["url"]
    quakes.append(event_dict)
```
